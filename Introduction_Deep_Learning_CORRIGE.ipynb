## âš™ï¸ Partie 2 â€” Chargement et visualisation des donnÃ©es

```python

(X_train, y_train), (X_test, y_test) = mnist.load_data()
print('Taille du jeu d\'entraÃ®nement :', X_train.shape)
print('Taille du jeu de test :', X_test.shape)

plt.figure(figsize=(8,4))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(f'Label: {y_train[i]}')
    plt.axis('off')
plt.show()
```

## ðŸ§© Partie 3 â€” PrÃ©traitement des donnÃ©es

```python

X_train = X_train / 255.0
X_test = X_test / 255.0

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

## ðŸ§  Partie 4 â€” CrÃ©ation du modÃ¨le de rÃ©seau de neurones

```python

model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.summary()
```

## âš™ï¸ Partie 5 â€” Compilation et entraÃ®nement du modÃ¨le

```python

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=5, batch_size=32, validation_split=0.1, verbose=1)
```

## ðŸ“Š Partie 6 â€” Ã‰valuation du modÃ¨le

```python

test_loss, test_acc = model.evaluate(X_test, y_test)
print(f'PrÃ©cision sur les donnÃ©es de test : {test_acc*100:.2f}%')
```

## ðŸ“ˆ Partie 7 â€” Visualisation des courbes d'apprentissage

```python

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('PrÃ©cision du modÃ¨le')
plt.xlabel('Ã‰pochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Fonction de perte')
plt.xlabel('Ã‰pochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

## ðŸ” Partie 8 â€” PrÃ©dictions et visualisation des rÃ©sultats

```python

predictions = model.predict(X_test)

plt.figure(figsize=(10,5))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_test[i], cmap='gray')
    plt.title(f'PrÃ©dit: {np.argmax(predictions[i])}')
    plt.axis('off')
plt.show()
```
