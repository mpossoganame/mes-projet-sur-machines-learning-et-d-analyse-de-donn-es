```python
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv('iris.csv')
```

```python
from re import X
# Entraîne ton modèle pour qu’il trouve 3 clusters à partir des 4 colonnes numériques.
from sklearn.cluster import KMeans

X = df.drop('species',axis = 1)

modelKM = KMeans(n_clusters=3)
modelKM.fit(X)
```

```python
df['model_label'] = modelKM.labels_
```

```python
cluster = modelKM.cluster_centers_
cluster[1,:]
```

```python
#Produis une visualisation comparant le dataset de base avec les classes générées par le K-means en t’inspirant de cette image.
#Tu peux prendre 2 dimensions de ton choix parmi les 4.
#Sur le graphique de gauche, tu dois afficher les points avec des couleurs différentes pour chaque cluster d'appartenance, ainsi que le centre de chaque cluster bien en évidence (ici les croix vertes).
#Sur le graphique de droite, tu dois afficher les points avec des couleurs différentes correspondant à leur espèce d'iris species

import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10, 5))

plt.subplot(121)

sns.scatterplot(data = df, x = 'sepal_length', y = 'petal_width', hue = 'model_label', palette = sns.color_palette("tab10", n_colors=3));
plt.scatter(cluster[:,0],cluster[:,3],color ='g',marker = 'x')

plt.legend(title='label',loc='upper left');
plt.title('3 clusters')



plt.subplot(122)

sns.scatterplot(data = df, x = 'sepal_length', y = 'petal_width', hue ='species', palette = sns.color_palette("tab10", n_colors=3));

plt.title('3 original species')
plt.legend(title='species',loc='upper left')


```

```python
#Entraine un autre modèle sur 2 clusters, puis affiche les mêmes graphiques
modelKM2 = KMeans(n_clusters=2)
modelKM2.fit(X)
df['model_label2'] = modelKM2.labels_
cluster2 = modelKM2.cluster_centers_

import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10, 5))

plt.subplot(121)

sns.scatterplot(data = df, x = 'sepal_length', y = 'petal_width', hue = 'model_label2', palette = sns.color_palette("tab10", n_colors=2));
plt.scatter(cluster2[:,0],cluster2[:,3],color ='g',marker = 'x')

plt.legend(title='label',loc='upper left');
plt.title('2 clusters')



plt.subplot(122)

sns.scatterplot(data = df, x = 'sepal_length', y = 'petal_width', hue ='species', palette = sns.color_palette("tab10", n_colors=3));

plt.title('3 original species')
plt.legend(title='species',loc='upper left')
```

```python
#Idem pour 4 clusters
modelKM4 = KMeans(n_clusters=4)
modelKM4.fit(X)
df['model_label4'] = modelKM4.labels_
cluster4 = modelKM4.cluster_centers_

import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10, 5))

plt.subplot(121)

sns.scatterplot(data = df, x = 'sepal_length', y = 'petal_width', hue = 'model_label4', palette = sns.color_palette("tab10", n_colors=4));
plt.scatter(cluster4[:,0],cluster4[:,3],color ='g',marker = 'x')

plt.legend(title='label',loc='upper left');
plt.title('4 clusters')



plt.subplot(122)

sns.scatterplot(data = df, x = 'sepal_length', y = 'petal_width', hue ='species', palette = sns.color_palette("tab10", n_colors=3));

plt.title('3 original species')
plt.legend(title='species',loc='upper left')
```

```python
from sklearn.metrics  import silhouette_score
scoreS=[]
for k in range(2,11):
    modelKM = KMeans(n_clusters=k)
    modelKM.fit(X)
    scoreS.append(silhouette_score(X, modelKM.labels_))
scoreS
plt.plot(range(2,11),scoreS)
```
